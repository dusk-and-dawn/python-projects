```python
def sentence_puzzle(incoming_string = input('tell me a sentence: ')):
    vowels = ['a', 'e', 'i', 'o', 'u']
    sentence_list = incoming_string.split(' ')
    final_list = []
    for word in sentence_list:
        if word[0] in vowels:
            x = str(word+'way')
            final_list.append(x)
        else:
            y = str(word[1:]+word[0]+'ay')
            final_list.append(y)
    for word in final_list: 
        print(word, end=' ')
sentence_puzzle()
```
